🧠 **Storytelling**
Relatório que mostre como o faturamento evoluiu com o tempo. O objetivo é identificar picos de vendas, momentos de queda e tendências de crescimento que ajudem a empresa a planejar melhor suas ações.

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar os dados
df = pd.read_excel('/kaggle/input/dataset/dados.xlsx')
```

```python
# Converter coluna de data para datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])
```

```python
# Criar colunas auxiliares
df['Ano-Mês'] = df['data_venda'].dt.to_period('M')
df['Ano-Semana'] = df['data_venda'].dt.to_period('W')
df['Data_Dia'] = df['data_venda'].dt.date
```

```python
# Agrupar o faturamento líquido por:
faturamento_dia = df.groupby('Data_Dia')['valor total liquido'].sum()
faturamento_semana = df.groupby('Ano-Semana')['valor total liquido'].sum()
faturamento_mes = df.groupby('Ano-Mês')['valor total liquido'].sum()
```

```python
# Plotar os gráficos
plt.figure(figsize=(16, 10))
```

```python
# 1. Faturamento Diário
plt.subplot(3, 1, 1)
faturamento_dia.plot(marker='o', color='gray')
plt.title('📅 Faturamento Diário')
plt.ylabel('R$')
plt.grid(True)
```

```python
# 2. Faturamento Semanal
plt.subplot(3, 1, 2)
faturamento_semana.plot(marker='o', color='orange')
plt.title('🗓️ Faturamento Semanal')
plt.ylabel('R$')
plt.grid(True)
```

```python
# 3. Faturamento Mensal
plt.subplot(3, 1, 3)
faturamento_mes.plot(marker='o', color='seagreen')
plt.title('📆 Faturamento Mensal')
plt.ylabel('R$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Mostrar os dados resumidos
print("📅 Faturamento Diário:\n", faturamento_dia.tail(), "\n")
print("🗓️ Faturamento Semanal:\n", faturamento_semana.tail(), "\n")
print("📆 Faturamento Mensal:\n", faturamento_mes.tail(), "\n")
```
