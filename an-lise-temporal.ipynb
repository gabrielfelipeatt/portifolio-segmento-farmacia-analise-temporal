```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar os dados
df = pd.read_excel('/kaggle/input/dataset/dados.xlsx')
```

```python
# Converter coluna de data para datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])
```

```python
# Criar colunas auxiliares
df['Ano-M√™s'] = df['data_venda'].dt.to_period('M')
df['Ano-Semana'] = df['data_venda'].dt.to_period('W')
df['Data_Dia'] = df['data_venda'].dt.date
```

```python
# Agrupar o faturamento l√≠quido por:
faturamento_dia = df.groupby('Data_Dia')['valor total liquido'].sum()
faturamento_semana = df.groupby('Ano-Semana')['valor total liquido'].sum()
faturamento_mes = df.groupby('Ano-M√™s')['valor total liquido'].sum()
```

```python
# Plotar os gr√°ficos
plt.figure(figsize=(16, 10))
```

```python
# 1. Faturamento Di√°rio
plt.subplot(3, 1, 1)
faturamento_dia.plot(marker='o', color='gray')
plt.title('üìÖ Faturamento Di√°rio')
plt.ylabel('R$')
plt.grid(True)
```

```python
# 2. Faturamento Semanal
plt.subplot(3, 1, 2)
faturamento_semana.plot(marker='o', color='orange')
plt.title('üóìÔ∏è Faturamento Semanal')
plt.ylabel('R$')
plt.grid(True)
```

```python
# 3. Faturamento Mensal
plt.subplot(3, 1, 3)
faturamento_mes.plot(marker='o', color='seagreen')
plt.title('üìÜ Faturamento Mensal')
plt.ylabel('R$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Mostrar os dados resumidos
print("üìÖ Faturamento Di√°rio:\n", faturamento_dia.tail(), "\n")
print("üóìÔ∏è Faturamento Semanal:\n", faturamento_semana.tail(), "\n")
print("üìÜ Faturamento Mensal:\n", faturamento_mes.tail(), "\n")
```
